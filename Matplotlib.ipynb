# xlabel, ylabel, title.

```python
x_val = np.arange(-2,2,step=0.1)
y_val = x_val**2
plt.plot(x_val,y_val)
plt.xlabel("x values")
plt.ylabel("square values")
plt.title("y=x**2")
plt.show()
```

# Dotted Line graph

```python
x_val = np.arange(-2,2,step=0.1)
y_val = x_val**2
plt.plot(x_val,y_val,"r*")
plt.xlabel("x values")
plt.ylabel("square values")
plt.title("y=x**2")
plt.show()
```

# Accessing the certain part of the graph(xlim,ylim)

```python
x_val = np.arange(-2,2,step=0.1)
y_val = x_val**2
plt.plot(x_val,y_val,"r*")
plt.xlabel("x values")
plt.ylabel("square values")
plt.title("y=x**2")
plt.xlim(left=0,right=2)
plt.show()
```

```python
x_val = np.arange(-2,2,step=0.1)
y_val = x_val**2
plt.plot(x_val,y_val,"r*")
plt.xlabel("x values")
plt.ylabel("square values")
plt.title("y=x**2")
plt.xlim(left=-2,right=0)
plt.show()
```

```python
x_val = np.arange(-2,2,step=0.1)
y_val = x_val**2
plt.plot(x_val,y_val,"r*")
plt.xlabel("x values")
plt.ylabel("square values")
plt.title("y=x**2")
plt.xlim(left=-2,right=0)
plt.ylim(bottom=0,top=2)
plt.show()
```

# Legends

```python
x = np.arange(-2,2,step=0.1)
y = x**2
y2 = x**3
plt.plot(x,y,label="Squares")
plt.plot(x,y2,label="Cubes")
plt.legend()
plt.show()
```

```python
x = np.arange(-2,2,step=0.1)
y = x**2
y2 = x**3
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(["squares","cubes"],loc = "lower right")
plt.show()
```

```python
x = np.arange(-2,2,step=0.1)
y = x**2
y2 = x**3
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(["squares","cubes"],loc = "lower center")
plt.show()
```

```python
x = np.arange(-2,2,step=0.1)
y = x**2
y2 = x**3
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(["squares","cubes"],loc = "best")
plt.show()
```

# Grid

```python
x = np.arange(-2,2,step=0.1)
y = x**2
y2 = x**3
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(["squares","cubes"],loc = "upper center")
plt.grid()
plt.show()
```

```python
x = np.arange(-2,2,step=0.1)
y = x**2
y2 = x**3
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(["squares","cubes"],loc = "best")
plt.text(0,0,"(0.0)")
plt.show()
```
