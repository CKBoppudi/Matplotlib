# Grid

```python
x = np.arange(-2,2,step=0.1)
y = x**2
y2 = x**3
plt.plot(x,y)
plt.plot(x,y2)
plt.legend(["squares","cubes"],loc = "upper center")
plt.grid()
plt.show()
```

```python
x = np.arange(-2,2,step=0.1)
y = x**2
y2 = x**3
y3 = x**4
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.legend(["squares","cubes","raised to the power 4"],loc = "lower right")
plt.text(0,0,"point")
plt.show()
```

# Sub plots.

```python
x  = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
fig , ax = plt.subplots(2,1,figsize=(10,10)) # 2,1 represents plots to be in 2 rows and 1 column 
ax[0].plot(x,y1)
ax[1].plot(x,y2)
plt.show()
```

```python
x  = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x)
y4 = np.tan(x)
fig , ax = plt.subplots(4,1 , figsize=(10,10)) # 4,1 represents the plots to be in 4 rows and one column.
ax[0].plot(x,y1)
ax[0].set_title("sinx")
ax[1].plot(x,y2)
ax[1].set_title("cosx")
ax[2].plot(x,y3)
ax[2].set_title("tanx")
ax[3].plot(x,y4)
ax[3].set_title("log")
plt.show()
```

```python
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log(x)
fig , ax = plt.subplots(2,2,figsize = (10,10))
ax[0,0].plot(x,y1)
ax[0,0].set_title("sinx")
ax[0,1].plot(x,y2)
ax[0,1].set_title("cosx")
ax[1,0].plot(x,y3)
ax[1,0].set_title("tanx")
ax[1,1].plot(x,y4)
ax[1,1].set_title("logx")
plt.show()
```

# Subplots - method2

```python
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log(x)
plt.figure(figsize = (10,10))
plt.subplot(4,1,1) # In 4,1,1 - (4,1) represents rows, columns and 1 represents the position.
plt.plot(x,y1)
plt.subplot(4,1,2)
plt.plot(x,y2)
plt.subplot(4,1,3)
plt.plot(x,y3)
plt.subplot(4,1,4)
plt.plot(x,y4)
plt.show()
```

```python
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log(x)
plt.figure(figsize = (10,10))
plt.subplot(2,2,1) # In 2,2,1 - (2,2) represents rows, columns and 1 represents the position.
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()
```
